```python
desc = reviews.description
first_discription = reviews.description.iloc[0] 
display(first_discription)
```

```python
first_row = reviews.iloc[0]
display(first_row)
```

```python
first_10_descriptions = reviews.description[0:10]
display(first_10_descriptions)
```

```python
#Select the records with index labels `1`, `2`, `3`, `5`, and `8`, assigning the result to the variable `sample_reviews`.
indices = [1, 2, 3, 4, 5, 8]
sample_reviews = reviews.loc[indices]
display(sample_reviews)
```

```python
#Create a variable `df` containing the `country`, `province`, `region_1`, and `region_2` columns of the records with the index labels `0`, `1`, `10`, and `100`. 
# In other words, generate the following `DataFrame`:
cols = ['country', 'province', 'region_1', 'region_2']
indices = [0, 1, 10, 100]
df = reviews.loc[indices, cols]
display(df)
```

```python
#Create a variable `df` containing the `country` and `variety` columns of the first 100 records
cols= ['country', 'variety']
df = reviews.loc[:99, cols]
display(df.head())
```

```python
#Create a DataFrame `italian_wines` containing reviews of wines made in `Italy`. Hint: `reviews.country` equals what?
italian_wines = reviews[reviews.country == 'Italy']
display(italian_wines)
```

```python
#Create a DataFrame `top_oceania_wines` containing all reviews with at least 95 points (out of 100) for wines from Australia or New Zealand.
top_oceania_wines = reviews.loc[
    (reviews.country.isin(['Australia', 'New Zeland']))
    & (reviews.points >= 95)
]
display(top_oceania_wines)
```

```python
## Plot bar charts for categorical data
reviews['province'].value_counts().head(10).plot.bar()
```

```python
#Percentage of wine that each category accounts for
(reviews['province'].value_counts().head(10) / len (reviews)).plot.bar()
```

```python
#Number of reviews of a certain score and plot as bar graph
reviews['points'].value_counts().sort_index().plot.bar()
```

```python
#using line chart to plot values
reviews['points'].value_counts().sort_index().plot.line()
```

```python
#Build area plot
reviews['points'].value_counts().sort_index().plot.area()
```

```python
#plot interval data using histograms
reviews[reviews['price'] < 200]['price'].plot.hist()
```

```python
#skewed data ploted on a histogram
reviews['price'].plot.hist()
```

```python
#plot of wine scores as histogram
reviews['points'].plot.hist()
```
